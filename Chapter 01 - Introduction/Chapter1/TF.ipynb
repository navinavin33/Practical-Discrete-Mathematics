```python
import pandas as pd
import itertools

import pandas as pd

def AND(a, b):
    return a and b

def OR(a, b):
    return a or b

def NOT(a):
    return not a
# Define the propositions
propositions = [True, False]

# Create an empty list to store the rows of the truth table
truth_table = []

# Generate the truth table
for P in propositions:
    for Q in propositions:
        truth_table.append({
            'P': P,
            'Q': Q,
            'P AND Q': AND(P, Q),
            'P OR Q': OR(P, Q),
            'NOT P': NOT(P),
            'NOT Q': NOT(Q)
        })

# Convert to DataFrame for better visualization
df = pd.DataFrame(truth_table)
print(df)

```

```python

import itertools
import pandas as pd
# Function to safely evaluate the proposition
def evaluate_proposition(expression, P, Q):
    # Use eval to evaluate the logical expression dynamically
    # Mapping 'and', 'or', 'not' for user-friendliness
    try:
        return eval(expression)
    except Exception as e:
        print(f"Error evaluating the expression: {e}")
        return None
# Function to create a truth table from a user-defined proposition
def generate_truth_table(expression):
    variables = ['P', 'Q']
    # Generate all possible combinations of truth values
    truth_values = list(itertools.product([True, False], repeat=len(variables)))
    # Create the truth table data
    truth_table_data = []
    for row in truth_values:
        P, Q = row
        result = evaluate_proposition(expression, P, Q)
        truth_table_data.append({
            'P': P,
            'Q': Q,
            'Result': result
        })
    # Create a DataFrame using pandas
    df = pd.DataFrame(truth_table_data)
    return df
# Get user input for logical expression
user_input = input("Enter a logical expression (e.g., P and Q or not P): ")
# Generate and display the truth table
truth_table = generate_truth_table(user_input)
truth_table

```

### The `shuffle()` function

```python
import random

# Set a random seed so the code is reproducible
random.seed(1)

# Run the random.shuffle() function 5 times and display the outputs
for i in range(0,5):
    numbers = [3, 1, 4, 12, 8, 5, 2, 9]
    random.shuffle(numbers)
    print(numbers)
```
